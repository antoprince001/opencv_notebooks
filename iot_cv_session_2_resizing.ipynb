# Image Resizing using OpenCV


* In this Jupyter notebook let's try to resize the image using OpenCV Library

```python
#let's start with the Imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Read the Image
* Read the image using **cv2.imread()** function
* Display the image using **plt.imshow()** function

* When **cv2.imread()** function reads an image the color scale is in **BGR** format so everytime we are displaying the image we will be converting it back to **RGB** format

```python
# Create a NumPy array with random data
width = 6
height = 6
channels = 3  # RGB image with 3 channels
dtype = np.uint8  # Data type for image pixels (8-bit unsigned integer)

# Generate random pixel values between 0 and 255
random_image_data = np.random.randint(0, 256, size=(height, width, channels), dtype=dtype)

# Convert the NumPy array to an image
image = cv2.cvtColor(random_image_data, cv2.COLOR_BGR2RGB)  # Convert RGB to BGR for OpenCV
```

```python
plt.imshow(image)
```

```python
image.shape
```

## Resize the Image (Make it small)
 * let's down size the image using new width and height

```python
down_Width = int(width/2)
down_Height = int(height/2)
down_Points = (down_Width, down_Height)
resized_Down = cv2.resize(image, down_Points, interpolation= cv2.INTER_LINEAR)
```

```python
plt.imshow(resized_Down)
```

## Resize the Image (Make it big)
 * let's up scale the image using new width and height

```python
up_Width = int(width*2)
up_Height = int(height*2)
up_Points = (up_Width, up_Height)
resized_Up = cv2.resize(image, up_Points, interpolation= cv2.INTER_LINEAR)
```

```python
plt.imshow(resized_Up)
```

```python
plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(image);plt.title("Original Image")
plt.subplot(132);plt.imshow(resized_Up);plt.title("Up Scaled Image")
plt.subplot(133);plt.imshow(resized_Down);plt.title("Down Sized Image")
# # Create a subplot with 1 row, 3 columns, and position 1
```

# Interpolation
* Let's try resizing with the images with different interpolation methods
* Different Interpolation methods are used for different functions   
 * **cv2.INTER_AREA** is best used for **shrinking** of Images because it resamples the image using pixel area relation as it gives more free results
 * **cv2.INTER_CUBIC** (slow) is  best used for **zooming** of the Image , it uses a bicubic interpolation over **4x4 pixel** neighborhood

 * **cv2.INTER_LINEAR**  uses a bilinear interpolation and is  used for all resizing purposes, hence it is used as **default** in many resizing functions .

 * If you prefer speed over quality use a **Nearest Neighbor method** but if you want to preserve the quality after downsampling then **interpolation** is a better option but it also depends on image content

```python
scale_up = 2.5
inter_Nearest = cv2.resize(image, None, fx= scale_up, fy= scale_up, interpolation= cv2.INTER_NEAREST)
inter_Linear = cv2.resize(image, None, fx= scale_up, fy= scale_up, interpolation= cv2.INTER_LINEAR)
inter_Area = cv2.resize(image, None, fx= scale_up, fy= scale_up, interpolation= cv2.INTER_AREA)
```

```python
plt.figure(figsize=[20,20])
plt.subplot(141)
plt.title("ORIGINAL")
plt.imshow(image)
plt.subplot(142)
plt.title("INTER_NEAREST")
plt.imshow(inter_Nearest)
plt.subplot(143)
plt.title("INTER_LINEAR")
plt.imshow(inter_Linear)
plt.subplot(144)
plt.title("INTER_AREA")
plt.imshow(inter_Area)
```

```python
scale_down = 0.75
inter_Nearest = cv2.resize(image, None, fx= scale_down, fy= scale_down, interpolation= cv2.INTER_NEAREST)
inter_Linear = cv2.resize(image, None, fx= scale_down, fy= scale_down, interpolation= cv2.INTER_LINEAR)
inter_Area = cv2.resize(image, None, fx= scale_down, fy= scale_down, interpolation= cv2.INTER_AREA)
```

```python
plt.figure(figsize=[20,20])
plt.subplot(141)
plt.title("ORIGINAL")
plt.imshow(image)
plt.subplot(142)
plt.title("INTER_NEAREST")
plt.imshow(inter_Nearest)
plt.subplot(143)
plt.title("INTER_LINEAR")
plt.imshow(inter_Linear)
plt.subplot(144)
plt.title("INTER_AREA")
plt.imshow(inter_Area)
```
